### Qual a semantica entre:
* DATA_INGRESSO_CARGOFUNCAO; 
* DATA_NOMEACAO_CARGOFUNCAO; 
* DATA_INGRESSO_ORGAO; 
* DATA_DIPLOMA_INGRESSO_SERVICOPUBLICO;

### Hipótese:
DATA_DIPLOMA_INGRESSO_SERVICOPUBLICO => DATA_INGRESSO_ORGAO => DATA_INGRESSO_CARGOFUNCAO
DATA_NOMEACAO_CARGOFUNCAO é inexistente.

Mapeamento no Portal da Tranparência
* DATA_DIPLOMA_INGRESSO_SERVICOPUBLICO => Ingresso no Serviço Público.Data de publicação
* DATA_INGRESSO_ORGAO => Data da última alteração no Órgão
* DATA_INGRESSO_CARGOFUNCAO => Data da Última Alteração no Cargo/Data da Última Alteração na Função
